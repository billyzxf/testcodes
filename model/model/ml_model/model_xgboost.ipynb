```python
label_data_new = label_data_new.fillna(0, subset=['regged', 'ordered', 'paid'])
```

```python
from pyspark.ml import Pipeline
```

```python
def save_pipeline_model(path):
    """
    Save the pipeline model to the specified path.
    """
    string_indexers = [E.StringIndexer(inputCol=col, outputCol=col + "_index", handleInvalid='keep') for col in categorical_features]
    one_hot_encoders = [E.OneHotEncoder(inputCols=[col + "_index"], outputCols=[col + "_ohe"], handleInvalid='keep') for col in categorical_features]
    tokenizers = [E.Tokenizer(inputCol=col, outputCol=col + "_token") for col in text_features]
    hashing_tfs = [E.HashingTF(inputCol=col + "_token", outputCol=col + "_hash", numFeatures=10000) for col in text_features]
    idfs = [E.IDF(inputCol=col + "_hash", outputCol=col + "_tfidf") for col in text_features]
    vector_assembler = E.VectorAssembler(
        inputCols=numerical_features + [col + "_ohe" for col in categorical_features] +
                  [col + "_tfidf" for col in text_features],
        outputCol="features",
        handleInvalid='keep'
    )
    pipeline = Pipeline(stages=string_indexers + one_hot_encoders + tokenizers + hashing_tfs + idfs + [vector_assembler])
    pipeline_model = pipeline.fit(label_data_new)
    print(f"Pipeline model saved to {path}")
    return pipeline_model
```

```python
pipeline_model = save_pipeline_model("model/pipeline_model")
processed_data = pipeline_model.transform(label_data_new)
```

```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import log_loss
from xgboost import XGBClassifier

def xgboost_grid_search(data, label_col='label', features_col='features'):
    pandas_df = data.select(label_col, features_col).toPandas()
    X = np.array(pandas_df[features_col].tolist())
    y = pandas_df[label_col].values

    param_grid = {
        'max_depth': [4, 6, 8],
        'learning_rate': [0.01, 0.1, 0.2],
        'n_estimators': [50, 100, 200],
        'subsample': [0.8, 1.0]
    }
    xgb_clf = XGBClassifier(objective='binary:logistic', eval_metric='logloss', use_label_encoder=False)
    grid_search = GridSearchCV(xgb_clf, param_grid, scoring='neg_log_loss', cv=3, verbose=1)
    grid_search.fit(X, y)
    print("Best parameters:", grid_search.best_params_)
    print("Best log loss:", -grid_search.best_score_)
    return grid_search.best_estimator_

def train_xgboost_model(data, label_col='label', features_col='features'):
    """
    Train an XGBoost model using the provided data.
    """
    # Convert Spark DataFrame to Pandas DataFrame
    pandas_df = data.select(label_col, features_col).toPandas()
    
    # Prepare the data for XGBoost
    X = np.array(pandas_df[features_col].tolist())
    y = pandas_df[label_col].values
    
    # Create DMatrix for XGBoost
    dtrain = xgb.DMatrix(X, label=y)
    
    # Set parameters for XGBoost
    params = {
        'objective': 'binary:logistic',
        'eval_metric': 'logloss',
        'max_depth': 6,
        'eta': 0.1,
        'seed': 42
    }
    
    # Train the model
    num_rounds = 100
    model = xgb.train(params, dtrain, num_rounds)
    
    print("XGBoost model trained successfully.")
    return model
```
